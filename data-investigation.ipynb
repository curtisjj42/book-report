```python
from ebooklib import epub
import ebooklib

book = epub.read_epub('Data/raw/1_Eragon_by_Paolini_Christopher.epub')

for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
    print(item.get_content().decode('utf-8'))
```

```python
from bs4 import BeautifulSoup

content = []
for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
    soup = BeautifulSoup(item.get_content(), 'html.parser')
    content.append(soup.get_text())
```

```python
full_text = "\n".join(content)
```

```python
full_text
```

```python
import nltk

nltk.download('punkt_tab')
```

```python
import re

# Basic cleaning
cleaned_text = re.sub(r'\s+', ' ', full_text)  # Remove extra whitespace
cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)  # Remove punctuation
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(cleaned_text)
```

```python
with open('Data/processed/Eragon_cleaned.txt', 'w', encoding='utf-8') as f:
    f.write(cleaned_text)
```

```python
chapters={}
for item in book.get_items_of_type(ebooklib.ITEM_DOCUMENT):
    soup = BeautifulSoup(item.get_content(), 'html.parser')
    chapter_title = soup.title.string if soup.title else f"Chapter {len(chapters)+1}"
    chapters[chapter_title] = soup.get_text()
```

```python
from collections import Counter
word_freq = Counter(tokens)
print(word_freq.most_common(10))
```
